```python
data = pd.read_csv("./Data_cleaned/Pop_BE_Changement_population_clean.csv", sep=";", )
data = data[data["Année de référence"] == "1992-01-01"]
```

```python
def INS_to_arrondissement (dataframe, INS_col, arrondissement_col):
    """
    Converts the column containing INS codes into the corresponding arrondissement, then adds arrondissement to the DataFrame.

        • dataframe: pandas.DataFrame item to which to add the column.
        • INS_col: str, name of column in your pandas.DataFrame containing the INS codes.
        • arrondissement_col: str, name to give to the new column that will contain arrondissements.
    """
    
    dataframe[arrondissement_col] = ""
    dataframe[INS_col] = dataframe[INS_col].astype(int)
    
    for INS in dataframe[INS_col]:
        if INS > 93000:
            dataframe[arrondissement_col] = "Philippeville"
        
        if INS > 92000:
            dataframe[arrondissement_col] = "Namur"
        
        if INS > 91000:
            dataframe[arrondissement_col] = "Dinant"
        
        if INS > 85000:
            dataframe[arrondissement_col] = "Virton"
        
        if INS > 84000:
            dataframe[arrondissement_col] = "Neufchâteau"
        
        if INS > 83000:
            dataframe[arrondissement_col] = "Marche-en-Famenne"
        
        if INS > 82000:
            dataframe[arrondissement_col] = "Bastogne"
        
        if INS > 81000:
            dataframe[arrondissement_col] = "Arlon" 
        
        if INS > 73000:
            dataframe[arrondissement_col] = "Tongres"
        
        if INS > 72000:
            dataframe[arrondissement_col] = "Maaseik"

        if INS > 71000:
            dataframe[arrondissement_col] = "Hasselt"

        if INS > 64000:
            dataframe[arrondissement_col] = "Waremme"

        if INS > 63000:
            dataframe[arrondissement_col] = "Verviers"

        if INS > 62000:
            dataframe[arrondissement_col] = "Liège"

        if INS > 61000:
            dataframe[arrondissement_col] = "Huy"

        if INS > 58000:
            dataframe[arrondissement_col] = "La Louvière"
  
        if INS > 57000:
            dataframe[arrondissement_col] = "Tournai"

        if INS > 56000:
            dataframe[arrondissement_col] = "Thuin"

        if INS > 55000:
            dataframe[arrondissement_col] = "Soignies"

        if INS > 54000:
            dataframe[arrondissement_col] = "Mouscron"
  
        if INS > 53000:
            dataframe[arrondissement_col] = "Mons"
   
        if INS > 52000:
            dataframe[arrondissement_col] = "Charleroi"
   
        if INS > 51000:
            dataframe[arrondissement_col] = "Ath"
    
        if INS > 46000:
            dataframe[arrondissement_col] = "Saint-Nicolas"
    
        if INS > 45000:
            dataframe[arrondissement_col] = "Audenarde"
    
        if INS > 44000:
            dataframe[arrondissement_col] = "Gand"
    
        if INS > 43000:
            dataframe[arrondissement_col] = "Eeklo"
    
        if INS > 42000:
            dataframe[arrondissement_col] =  "Termonde"

        if INS > 41000:
            dataframe[arrondissement_col] = "Alost"
 
        if INS > 38000:
            dataframe[arrondissement_col] = "Furnes"
 
        if INS > 37000:
            dataframe[arrondissement_col] = "Tielt"
  
        if INS > 36000:
            dataframe[arrondissement_col] = "Roulers"
 
        if INS > 35000:
            dataframe[arrondissement_col] = "Ostende"
  
        if INS > 34000:
            dataframe[arrondissement_col] = "Courtrai"
   
        if INS > 33000:
            dataframe[arrondissement_col] = "Ypres"
    
        if INS > 32000:
            dataframe[arrondissement_col] = "Dixmude"
    
        if INS > 31000:
            dataframe[arrondissement_col] = "Bruges"
  
        if INS > 25000:
            dataframe[arrondissement_col] = "Nivelles" 
   
        if INS > 24000:
            dataframe[arrondissement_col] = "Louvain"
    
        if INS > 23000:
            dataframe[arrondissement_col] = "Hal-Vilvorde"
    
        if INS > 21000:
            dataframe[arrondissement_col] = "Bruxelles"
    
        if INS > 13000:
            dataframe[arrondissement_col] = "Turnhout"
  
        if INS > 12000:
            dataframe[arrondissement_col] = "Malines"
  
        if INS > 11000:
            dataframe[arrondissement_col] = "Anvers"
        
        else:
            dataframe[arrondissement_col] = "Inconnu"

```
