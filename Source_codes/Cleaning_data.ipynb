# Unpivot main fact table from SQL Server

```python
import numpy as np
import pandas as pd
import os
import json
import pyodbc
from sqlalchemy import create_engine, text
from sqlalchemy.engine import URL
```

```python
user_pass = pd.read_csv("../05_Librairies_python/04_SQL_librairies/SQL_pass.csv")
user_pass
```

```python
# Connection server
server = user_pass.at[0,"server"]
user_name = user_pass.at[0,"user"]
password = user_pass.at[0,"mdp"]

connect_string = ('DRIVER={SQL Server};SERVER='
                     +server
                     +';Trusted_Connection=yes;UID='
                     +user_name
                     +';PWD='
                     + password
                    )
```

```python
# Connection server + DB
server = user_pass.at[0,"server"]
username = user_pass.at[0,"user"]
password = user_pass.at[0,"mdp"]
database = "BelgiumPopulation"

connect_string = 'DRIVER={SQL Server};SERVER='+server+";DATABASE="+database+';Trusted_Connection=yes;UID='+username+';PWD='+ password
```

```python
### use pyodbc with sqlalchemy
connection_url = URL.create("mssql+pyodbc", query={"odbc_connect": connect_string})
engine = create_engine(connection_url) #+-= cnxn= pyodbc.connect(connection_string)
```

```python
# Import table

DBBel = """
            SELECT * 
            FROM BEPopNationalities
        """

# Import all tables
mytable = pd.read_sql_query(DBBel, con=engine)
```

```python
mytable
```

```python
mytable["INS-NIS"] = mytable["INS-NIS"].str.replace(pat="\n", repl="")
```

```python
mytable
```

```python
# Strip whitespaces
df_obj = mytable.select_dtypes("object")
print(df_obj)
```

```python
mytable[df_obj.columns] = df_obj.apply(lambda x: x.str.strip())
```

```python
mytable.iloc[:,6:].columns
```

```python
# Unpivot table
df = mytable.melt(id_vars=mytable.iloc[:,:6].columns, value_vars=mytable.iloc[:,6:].columns)
```

```python
df
```

```python
df["variable"] = df["variable"].str.strip()
```

```python
df = df.rename(mapper={"variable":"pays", "value":"population"}, axis=1)
df
```

```python
# Save as CSV
df.to_csv("./Data_cleaned/BEPop_commune-origine-sexe_1992-2024.csv", index=False, encoding= "utf-8-sig")
```
