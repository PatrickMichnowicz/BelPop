# Imports

```python
import numpy as np
import pandas as pd
import requests as req
```

```python
response_sources = req.get("https://bestat.statbel.fgov.be/bestat/api/datasources/")
print(response_sources)
```

```python
response_views = req.get("https://bestat.statbel.fgov.be/bestat/api/views/")
print(response_views)
```

```python
all_sources = pd.json_normalize(response_sources.json())
all_sources
```

```python
all_views = pd.json_normalize(response_views.json())
all_views
```

```python
all_views.to_excel("./API_views.xlsx")
```
